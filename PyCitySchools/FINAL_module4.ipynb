```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path

# File to Load 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
# Calculate the total number of unique schools
unique_school_list = school_data_complete["school_name"].unique()
school_count = len(unique_school_list)
school_count
```

```python
# Calculate the total number of students
student_count = school_data_complete["student_name"].count()
student_count
```

```python
# Calculate the total budget
# for sum of all values in column "budget": total_budget = school_data_complete["budget"].sum()
# for sum of unique budget values:
total_budget_unique = school_data_complete["budget"].unique()
total_budget = total_budget_unique.sum()
total_budget
```

```python
# Calculate the average (mean) math score
# with unique values
unique_average_math_score = school_data_complete["math_score"].unique()
unique_average_math_score = unique_average_math_score.mean()
unique_average_math_score
# with all values
average_math_score = school_data_complete["math_score"].mean()
average_math_score
```

```python
# Calculate the average (mean) reading score

# with unique values
unique_average_reading_score = school_data_complete["reading_score"].unique()
unique_average_reading_score = unique_average_reading_score.mean()
unique_average_reading_score
# with all values
average_reading_score = school_data_complete["reading_score"].mean()
average_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = school_data_complete[(school_data_complete["math_score"] >= 70)].count()["student_name"]
passing_math_percentage = passing_math_count / float(student_count) * 100
passing_math_percentage
```

```python
# Calculate the percentage of students who passed reading (hint: look at how the math percentage was calculated)  
passing_reading_count = school_data_complete[(school_data_complete["reading_score"] >= 70)].count()["student_name"]
passing_reading_percentage = passing_reading_count / float(student_count) * 100
passing_reading_percentage
```

```python
# Use the following to calculate the percentage of students that passed math and reading
passing_math_reading_count = school_data_complete[
    (school_data_complete["math_score"] >= 70) & (school_data_complete["reading_score"] >= 70)
].count()["student_name"]
overall_passing_rate = passing_math_reading_count /  float(student_count) * 100
overall_passing_rate
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame

dict = {'Total Schools': school_count, 'Total Students': student_count, 'Total Budget': total_budget, 'Average Math Score': average_math_score, 'Average Reading Score': average_reading_score, '% Passing Math': passing_math_percentage, '% Passing Reading': passing_reading_percentage, '% Overall Passing': overall_passing_rate} 
district_summary = pd.DataFrame(dict, index=([0]))

# Formatting
district_summary["Total Schools"] = district_summary["Total Schools"].map("{:,.1f}".format)
district_summary["Total Students"] = district_summary["Total Students"].map("{:,.2f}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,.3f}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.4f}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.5f}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.6f}".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.7f}".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.8f}".format)

# add title 'DISTRICT SUMMARY' to df
district_summary = district_summary.style.set_caption('DISTRICT SUMMARY')

# Display the DataFrame
district_summary
```

```python
# Calculate the average test scores per school

# Create empty lists to store calculated values
per_school_math = []
per_school_reading = []

# Iterate through each school
for school in school_data_complete["school_name"].unique():
    # Filter rows for the current school
    school_data_subset = school_data_complete[school_data_complete["school_name"] == school]
    
    # Calculate average math and reading scores for the current school
    average_math_score = school_data_subset["math_score"].mean()
    average_reading_score = school_data_subset["reading_score"].mean()

    # Append calculated values to the lists
    per_school_math.append(average_math_score)
    per_school_reading.append(average_reading_score)

# Add new columns to school_data_expanded with the calculated values
school_data_expanded["Per School Av. Math Score"] = per_school_math
school_data_expanded["Per School Av. Reading Score"] = per_school_reading
```

```python
# Calculate the %passing math %passing reading % overall passing

# Create empty lists to store calculated values
per_school_percent_passing_math = []
per_school_percent_passing_reading = []
per_school_percent_passing_overall = []

# Iterate through each school
for school in school_data_complete["school_name"].unique():
    # Filter rows for the current school
    school_data_subset = school_data_complete[school_data_complete["school_name"] == school]
    
    # Calculate %passing math %passing reading % overall passing for the current school
    percent_passing_math = ((school_data_subset[(school_data_subset["math_score"]>= 70)]["student_name"].count()) / school_data_subset["size"].unique()) * 100
    percent_passing_reading = ((school_data_subset[(school_data_subset["reading_score"]>= 70)]["student_name"].count()) / school_data_subset["size"].unique()) * 100
    percent_passing_both = ((school_data_subset[(school_data_subset["math_score"]>= 70) & (school_data_subset["reading_score"]>= 70)]["student_name"].count()) / school_data_subset["size"].unique()) * 100

    # Append calculated values to the lists
    per_school_percent_passing_math.append(percent_passing_math[0])
    per_school_percent_passing_reading.append(percent_passing_reading[0])
    per_school_percent_passing_overall.append(percent_passing_both[0])


# Add new columns to school_data_expanded with the calculated values
school_data_expanded["% Passing Math"] = per_school_percent_passing_math
school_data_expanded["% Passing Reading"] = per_school_percent_passing_reading
school_data_expanded["% Passing Overall"] = per_school_percent_passing_overall

school_data_expanded
```

```python
# Calculate the number of schools with math scores of 70 or higher (2 points)
schools_passing_math_count = school_data_expanded['Per School Av. Math Score'].count()
schools_passing_math_count = school_data_expanded[(school_data_expanded['Per School Av. Math Score'] >= 70)]["school_name"].count()

# Calculate the number of schools with reading scores of 70 or higher (2 points)
schools_passing_reading_count = school_data_expanded['Per School Av. Reading Score'].count()
schools_passing_reading_count = school_data_expanded[(school_data_expanded['Per School Av. Reading Score'] >= 70)]["school_name"].count()

# Use the provided code to calculate the schools that passed both math and reading with scores of 70 or higher (2 points)
schools_passing_math_and_reading = school_data_expanded[(school_data_expanded['Per School Av. Math Score'] >= 70) & (school_data_expanded['Per School Av. Reading Score'] >= 70)]["school_name"] # in a list
```

## Scores by School Spending

```python
import warnings

# Suppress FutureWarnings
warnings.simplefilter(action='ignore', category=FutureWarning)
# Establish the bins 
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]

# Create a copy of the school summary since it has the "Per Student Budget" 
school_spending_df = per_school_summary.copy()

# Convert "Per Student Budget" to numeric
school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"], errors='coerce')

# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Spending Ranges (Per Student)"] = pd.cut(school_spending_df["Per Student Budget"], bins=spending_bins, labels=labels, include_lowest=True)

# Calculate the averages
spending_math_scores = school_spending_df.groupby("Spending Ranges (Per Student)")["Average Math Score"].mean()
spending_reading_scores = school_spending_df.groupby("Spending Ranges (Per Student)")["Average Reading Score"].mean()
spending_passing_math = school_spending_df.groupby("Spending Ranges (Per Student)")["% Passing Math"].mean()
spending_passing_reading = school_spending_df.groupby("Spending Ranges (Per Student)")["% Passing Reading"].mean()
overall_passing_spending = school_spending_df.groupby("Spending Ranges (Per Student)")["% Overall Passing"].mean()

# Create the spending_summary DataFrame using the binned and averaged spending data
spending_summary = pd.DataFrame({"Average Math Score": spending_math_scores, "Average Reading Score": spending_reading_scores, "% Passing Math": spending_passing_math, "% Passing Reading": spending_passing_reading, "% Overall Passing": overall_passing_spending})

# Display the spending_summary DataFrame
spending_summary
```
